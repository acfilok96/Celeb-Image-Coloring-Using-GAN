```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
loss_rec = np.load("PixToPixGAN_Loss_Record.npz")
```

```python
dis_real = np.array(loss_rec['a'])
dis_fake = np.array(loss_rec['b'])
gen = np.array(loss_rec['c'])

print(dis_real.shape, dis_fake.shape, gen.shape)
```

```python
plt.figure(figsize=(25,6))
plt.plot(dis_real,color='r')
plt.ylabel("Loss")  
plt.xlabel("Iteration")  
plt.title("Discriminator Loss Graph For Original Data")
plt.savefig("Discriminator Loss Graph For Original Data.png")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(25,6))
plt.plot(dis_fake,color='r')
plt.ylabel("Loss")  
plt.xlabel("Iteration")  
plt.title("Discriminator Loss Graph For Fake Data")
plt.savefig("Discriminator Loss Graph For Fake Data.png")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(25,6))
plt.plot(gen,color='r')
plt.ylabel("Loss")  
plt.xlabel("Iteration")  
plt.title("Generator Loss Graph")
plt.savefig("Generator Loss Graph.png")
plt.grid(True)
plt.show()
```
